```python
def ME(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.mean(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
def SD(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.std(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
```
