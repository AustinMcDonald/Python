```python
def ROI(arry):
    LE=len(arry)
    AA = np.array([])
    for x in range(0,LE):
        Eval = arry[x][0]
        if Eval>2.448 and Eval<2.477:
            AA = np.append(AA,Eval)
    return AA
    
def ROIB(arry):
    LE=len(arry)
    AA = np.array([])
    BB = np.array([])
    for x in range(0,LE):
        Eval = arry[x]
        if Eval[0]>2.448 and Eval[0]<2.477:
            AA = np.append(AA,Eval[1])
            BB = np.append(BB,Eval[2])
    return AA,BB
```
