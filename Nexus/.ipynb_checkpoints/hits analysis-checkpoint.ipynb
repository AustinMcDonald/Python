```python
data = pd.read_csv('hits.txt',names=['a'])
HitMap = pd.read_csv('channelposition.csv',names=['ID','detector','a','x','y'])
```

```python
hold = []
for x in range(0,len(data['a'])):
    aa=data['a'][x]
    bb=aa.split()
    hold.append(bb)
    
DataM = pd.DataFrame(hold,columns=['counter','detector','ID','x','y','z','E'])
    
```

```python
Data=DataM.copy()
Data['counter']= pd.to_numeric(Data['counter'],errors='coerce')
Data['ID']= pd.to_numeric(Data['ID'],errors='coerce')
Data['x']= pd.to_numeric(Data['x'],errors='coerce')
Data['y']= pd.to_numeric(Data['y'],errors='coerce')
Data['z']= pd.to_numeric(Data['z'],errors='coerce')
Data['E']= pd.to_numeric(Data['E'],errors='coerce')

Data = Data.dropna()
Data.loc[Data['counter']==32]
```

```python
event = dict()
eventPMT = dict()
eventSipm= dict()
for x in range(0,2):
    event[x] = Data.loc[Data['counter']==x].copy()
    eventSipm[x] = event[x].loc[event[x]['detector']=='SiPM']
    eventPMT[x]  = event[x].loc[event[x]['detector']=='PmtR11410']
```

```python
HitMapPMT  = HitMap.loc[HitMap['detector']=='PMTR11410']
HitMapSipm = HitMap.loc[HitMap['detector']=='SiPM50']
```

```python
plt.figure(figsize=(7,7))
plt.scatter(HitMapSipm['x'],HitMapSipm['y'],color=['b'],s=2)
plt.scatter(HitMapPMT['x'],HitMapPMT['y'],color=['r'],s=20)
plt.show()
```

```python
plt.figure(figsize=(14,14))
plt.subplot(221)
plt.scatter(eventSipm[0]['x'],eventSipm[0]['y'],c=eventSipm[0]['E'])
plt.xlim(-460,460)
plt.ylim(-460,460)

plt.subplot(222)
plt.scatter(eventPMT[0]['x'],eventPMT[0]['y'],c=eventPMT[0]['E'],s=1150)
plt.xlim(-460,460)
plt.ylim(-460,460)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(eventPMT[0]['x'],eventPMT[0]['y'],c=eventPMT[0]['E'],s=1150)
plt.xlim(-500,500)
plt.ylim(-500,500)
plt.show()
```
