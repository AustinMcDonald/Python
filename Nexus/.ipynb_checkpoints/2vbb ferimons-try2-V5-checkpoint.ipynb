```python
Q = 2.4578
Xval = np.linspace(0, Q, 1000)

def D(E):
    return E*(Q-E)**5*(1+2*E+4/3*E**2+1/3*E**3+1/30*E**4)
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))/(sig*np.sqrt(2*3.142))
```

```python
def Fermi(E,Per):
    E = E+.166*(1-Per)
    A = E*(Q-E)**5*(1+2*E+4/3*E**2+1/3*E**3+1/30*E**4)
    return A
```

```python
def LogL(Obs,Pre):
    A = (Obs * np.log(Pre))
    B = Pre
    #C = np.log(fact(Obs))
    LLH = sum(np.nan_to_num(A-B))#-C
    return -LLH
    
def LogLL(Obs,Pre):
    le = len(Obs)
    llh = np.array([])
    for x in range(0,le):
        A = (Obs[x] * np.log(Pre[x]))
        B = Pre[x]
        C = 0#math.log(math.factorial((Obs[x])))
        llh = np.append(llh,[A-B-C])
    return -llh

def LogLLL(observed,predicted):
    #KillBins = np.array([22, 23, 24, 25, 26, 27, 28])#np.where(predicted == 0)[0]
    KillBins = np.where(predicted == 0)[0]

    Obs = np.delete(observed,KillBins)
    Pre = np.delete(predicted,KillBins)
    A = (Obs * np.log(Pre))
    B = Pre
    #C = np.log(fact(Obs))
    LLH = np.sum(A-B)#-C
    return -LLH
```

```python
def smearing(Arry,sigma, size=1):
    LL = len(Arry)
    Out = np.array([])
    for x in range(0,LL):
        Point = Arry[x]
        Sigma = sigma[x]
        smr   = np.random.normal(Point, sigma, size)
        Out = np.append(Out,smr)
    return Out
```

```python
#X=Te  Y=D(Te)
def Con(X,Y,N,ffit):
    L = len(X)
    chunks = L/N
    XX = np.array_split(X,N)
    YY = np.array_split(Y,N)
    FF = np.array_split(ffit,N)
    ANS = np.array([])
    for q in range(1,N-1):
        xs = XX[q]
        ys = YY[q] 
        fs = FF[q]
        M = np.mean(xs)
        result = np.convolve(ys, gaussian(xs,M,fs), mode="full")/sum(gaussian(xs,M,fs))
        ugg = len(result)
        result = result[0:ugg/2+1]
        ANS = np.append(ANS,result)
    return ANS
```

```python
#X=Te  Y=D(Te)
def Con(X,Y,N,ffit):
    L = len(X)
    chunks = L/N
    XX = np.array_split(X,N)
    YY = np.array_split(Y,N)
    FF = np.array_split(ffit,N)
    ANS = np.array([])
    for q in range(1,N-1):
        L1 = len(XX[q-1])
        L2 = len(XX[q])
        L3 = len(XX[q+1])
        xs = np.append(XX[q-1],XX[q])
        xs = np.append(xs,XX[q+1])
        ys = np.append(YY[q-1],YY[q])
        ys = np.append(ys,YY[q+1])
        fs = np.append(FF[q-1],FF[q])
        fs = np.append(fs,FF[q+1])
        M = np.mean(xs)
        result = np.convolve(ys, gaussian(xs,M,fs), mode="full")/sum(gaussian(xs,M,fs))
        ugg = len(result)
        result = result[2*L1:2*L1+L2+1]
        #result = result[0:ugg/2]
        ANS = np.append(ANS,result)
    return ANS
```
