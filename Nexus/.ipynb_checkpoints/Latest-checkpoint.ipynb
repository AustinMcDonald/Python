```python
bb = np.genfromtxt('final-bb0nu.txt', delimiter=',')
bi = np.genfromtxt('final-bi214.txt', delimiter=',')
tl = np.genfromtxt('final-tl208.txt', delimiter=',')
```

```python
def ROI(arry):
    LE=len(arry)
    AA = np.array([])
    for x in range(0,LE):
        Eval = arry[x]
        if Eval[0]>2.448 and Eval[0]<2.477:
            AA = np.append(AA,[Eval])
    AA.reshape((LE, 3))
    return AA
    
```

```python
ROI(bb)
```
