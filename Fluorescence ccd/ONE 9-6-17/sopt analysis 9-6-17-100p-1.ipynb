```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import pandas as pd
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter

%matplotlib inline
from PIL import Image 
```

```python
files = os.listdir()
#files = files[1:] # gets rid of ds.store
#files
```

```python
a = (387,209)
b = (381,237)
c = (426,299)
d = (364,219)
#e = (428,251)
#f = (374,212)
Pairs = [a,b,c,d]
Pairs
```

```python
Row = 3
Col = 3
bb  = 10
Spots = dict()
BackG = dict()
for y in range(0,len(Pairs)):
    Xindex = Pairs[y][0]
    Yindex = Pairs[y][1]
    areaS = []
    areaB = []
    for x in range(0,len(files)):
        HOLD = Image.open(files[x])
        areaS.append(np.array(HOLD)[Yindex-Row:Yindex+Row+1,Xindex-Col:Xindex+Col+1].sum())
        areaB.append(np.array(HOLD)[Yindex+bb-Row:Yindex+bb+Row+1,Xindex-Col:Xindex+Col+1].sum())
    Spots[y] = areaS
    BackG[y] = areaB
```

```python
vals =np.arange(0,len(files))
for x in range(0,len(Pairs)):
    plt.figure(figsize=(15,8))
    plt.title(Pairs[x],fontsize=20)
    plt.scatter(vals,Spots[x],color='r')
    plt.scatter(vals,BackG[x])
    plt.tight_layout()
    plt.show()
#plt.savefig('/Users/austinmcdonald/Desktop/2Step2.pdf')
```

```python
svals = savgol_filter(vals, 21, 3)
sSpots = savgol_filter(Spots[0], 21, 5)
sBackG = savgol_filter(BackG[0], 21, 5)
plt.figure(figsize=(15,8))
plt.scatter(vals,Spots[0],color='r', alpha=0.4)
plt.scatter(vals,BackG[0],color='k', alpha=0.4)

plt.plot(svals,sSpots,color='r',linewidth=4)
plt.plot(svals,sBackG,color='k',linewidth=4)

plt.tight_layout()
#plt.xlim(100,900)
plt.show()
#plt.savefig('/Users/austinmcdonald/Desktop/2Step2.pdf')
```

```python
svals = savgol_filter(vals, 21, 3)
sSpots = savgol_filter(Spots[1], 21, 5)
sBackG = savgol_filter(BackG[1], 21, 5)
plt.figure(figsize=(15,8))
plt.scatter(vals,Spots[1],color='r', alpha=0.4)
plt.scatter(vals,BackG[1],color='k', alpha=0.4)

plt.plot(svals,sSpots,color='r',linewidth=4)
plt.plot(svals,sBackG,color='k',linewidth=4)

plt.tight_layout()
plt.xlim(100,900)
#plt.savefig('/Users/austinmcdonald/Desktop/9617-100p-4-409x208.pdf')
plt.show()
```
