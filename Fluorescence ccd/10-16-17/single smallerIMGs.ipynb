```python
xvs = np.arange(0,512)
Row = int(110) # define the area to take around the spots
Col = int(110) # define the area to take around the spots
Yindex = int(285)
Xindex = int(400)
for f in range(0,len(DataSet)):
    datas = WorkPath+'/'+DataSet[f]
    slides = os.listdir(datas)
    slides = [x for x in slides if x != '.DS_Store']
    for q in range(0,len(slides)):
        slide = datas+'/'+slides[q]
        Spots = os.listdir(slide)
        Spots = [x for x in Spots if x != '.DS_Store']
        for w in range(0,len(Spots)):
            dataPath = slide+'/'+Spots[w]
            files = os.listdir(dataPath)
            files = [x for x in files if x != '.DS_Store']
            CutFolder = '/Slide'+str(q)+'Spot'+str(w)
            os.mkdir(SavePath+'/'+ DataSet[f]+CutFolder)
            for s in range(0,len(files)):
                spot = np.array(Image.open(dataPath+'/'+files[s]))[Yindex-Row:Yindex+Row+1,Xindex-Col:Xindex+Col+1]
                im = Image.fromarray(spot.astype('float64'))
                im.save(SavePath+'/'+DataSet[f]+CutFolder +'/' + files[s])
                Images = None
```
