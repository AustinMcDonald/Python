```python
def ME(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.mean(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
def SD(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.std(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
```

```python
def findr(Arry,val):
    LEN = len(Arry)
    Out = np.zeros(LEN)
    LPs = np.where(Arry<val)
    RPs = np.where(Arry>=val)
    Out[RPs] = 1
    Uplim = Out
    Out = np.zeros(LEN)
    Out[LPs] = 1
    Lolim = Out
    return Lolim,Uplim
```
