```python
# volume to surface is maximum at z=2*r
R = 8.314
T = 300
def radius(P,n):
    return ((n*R*T)/(P*6.28))**(1/3)
def radiusS(P,n):
    return ((n*R*T)/(P*12.56))**(1/3)
```

```python
#1ton 7352.94 mol
mol = 7.35294
masses = np.arange(1,2000,1)*mol
masses
```

```python
plt.figure(figsize=(8,8))
plt.plot(masses/mol,radius(1.5e6,masses),linewidth=2,label='15 bar')
plt.plot(masses/mol,radius(1e6,masses),linewidth=2,label='10 bar')
plt.plot(masses/mol,radius(5e5,masses),linewidth=2,label='5 bar')

plt.plot(masses/mol,radiusS(1.5e6,masses),linewidth=2,linestyle='--',label='15 bar')
plt.plot(masses/mol,radiusS(1e6,masses),linewidth=2,linestyle='--',label='10 bar')
plt.plot(masses/mol,radiusS(5e5,masses),linewidth=2,linestyle='--',label='5 bar')

plt.xlabel(r'Mass [kg]',fontsize=26)
plt.ylabel(r'Radius [m]',fontsize=26)
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=5, width=1, which='minor')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='upper left',fontsize=24,shadow=True, fancybox=True)
plt.tight_layout()
plt.grid()
#plt.savefig('/Users/austinmcdonald/Desktop/positronoverlap.png')
plt.show()
```
