```python
files = os.listdir()
name=[]
for x in range(0,len(files)):
    name.append(files[x].replace('.txt',''))
```

```python

data = dict()
for x in range(0,len(files)):
    holder = pd.read_table(files[x],skiprows=20,names=['data']) 
    data[x] = pd.DataFrame(holder.data.str.split(',',1).tolist(),columns = ['amu','torr'])
    data[x]['amu'] = pd.to_numeric(data[x]['amu'], errors='coerce')
    data[x]['torr'] = pd.to_numeric(data[x]['torr'].str.replace(',', ''), errors='coerce')
```

```python
plt.figure(figsize=(15,9))
#for x in range(0,len(files)):
for x in range(0,2):
    plt.semilogy(data[x]['amu'],data[x]['torr'],linewidth=2,label=name[x],color=cm.hot((x)/3))
#plt.semilogy(amu2,torr2,color='steelblue',linewidth=2,label='Day2-Baking')
plt.xlabel('Mass / amu',fontsize=26)

plt.ylabel('Torr',fontsize=26)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xticks(np.arange(0, 51, 5.0))
plt.xlim(0,50)
plt.ylim(4e-10,3e-6)
plt.grid()
plt.legend(loc='upper left',fontsize=20)
plt.tight_layout()
#plt.savefig('Rga-5-8.pdf')

plt.show()
```
