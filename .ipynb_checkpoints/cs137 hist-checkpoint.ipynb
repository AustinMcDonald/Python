```python
#    Pulse hight histogram 
#        Have the data ready to feed in, just change the names
#        Pulseht1 and PulsehtBG.
#
#

hist, bin_edges = np.histogram(Pulseht1,bins=4000)     #makes the histogram
bin_c = (bin_edges[:-1]+bin_edges[1:])/2               #finds the centers of the bins

hist1, bin_edges1 = np.histogram(PulsehtBG,bins=4000)
bin_c1 = (bin_edges1[:-1]+bin_edges1[1:])/2

lhist = []                                             #defins the list for the loop
lbin_c = []

lhist1 = []
lbin_c1 = []
for q in range(0,len(hist)):                           #loops over the array to make list
    lhist.append(hist[q])
    lbin_c.append(bin_c[q])

for w in range(0,len(hist1)):
    lhist1.append(hist1[w])
    lbin_c1.append(bin_c1[w])
    
    
    
ph = list(zip(lbin_c,lhist))          #takes the list and makes them ordered pairs
ph1 = list(zip(lbin_c1,lhist1))

ph = [t for t in ph if t[1] > 0]     #filters all of the zeros out 
ph1 = [p for p in ph1 if p[1] > 0]

phx = np.array(ph)[:,0]              # makes the filtered list an array and slects x vals    
phy = np.array(ph)[:,1]              #slects yvals
phys = savgol_filter(phy, 21, 3)     #smooths yvals

ph1x = np.array(ph1)[:,0]
ph1y = np.array(ph1)[:,1]
ph1ys = savgol_filter(ph1y, 21, 3)


plt.figure(figsize=(8,8))
plt.xlim(0,-.8)
plt.plot(phx,phy,color='r',alpha=.4)
plt.plot(phx,phys,color='r',label='Cs137')
plt.plot(ph1x,ph1y,color='g',alpha=.4)
plt.plot(ph1x,ph1ys,color='g',label='background')

plt.legend(bbox_to_anchor=(.681, 1), loc=2, borderaxespad=0.,fontsize=18)

plt.axvline(-.4, color='b', linestyle=':')
plt.text(-.2,200,'Compton edge 470Kev',rotation=0,fontsize=15,color='b')
plt.axvline(-.54, color='k', linestyle=':')
plt.text(-.56,75,'Cs137 662Kev',rotation=0,fontsize=15,color='k')

plt.xlabel("Pulse Height", fontsize= 18)
plt.ylabel("No. of Events", fontsize= 18)
plt.suptitle("Pulse Height Distribution", size=22)
plt.show()
```

```python
#################################################################
#                                                               #
# This cell relys on the arrays defined above phx,phy,ph1x,ph1y #
#                                                               #
#################################################################

#
#  integral for hist
#
dxlist = []                  #defines list
dylist = []

for l in range(0,len(phx)):
    dxlist.append(phx[l])    # turns the arrays into list
    dylist.append(phy[l])
    
del dylist[len(dylist)-1]    # drop the last y val so the index matches

dx = diff(dxlist)            # find all the dx values

rects = dylist*dx            # gives the area of the rectangles
integral = sum(rects)        # adds them up 
print(integral)

#################################################
#  integral for hist1
#################################################
dxlist1 = []                  #defines list
dylist1 = []

for m in range(0,len(ph1x)):
    dxlist1.append(ph1x[m])    # turns the arrays into list
    dylist1.append(ph1y[m])
    
del dylist1[len(dylist1)-1]    # drop the last y val so the index matches

dx1 = diff(dxlist1)            # find all the dx values

rects1 = dylist1*dx1            # gives the area of the rectangles
integral1 = sum(rects1)        # adds them up 
print(integral1)
```

```python
#    Area histogram 
#        Have the data ready to feed in, just change the names
#        Area1 and AreaBG.
#
#

hist3, bin_edges3 = np.histogram(Area1,bins=200)     #makes the histogram
bin_c3 = (bin_edges3[:-1]+bin_edges3[1:])/2               #finds the centers of the bins

hist4, bin_edges4 = np.histogram(AreaBG,bins=200)
bin_c4 = (bin_edges4[:-1]+bin_edges4[1:])/2

lhist3 = []                                             #defins the list for the loop
lbin_c3 = []

lhist4 = []
lbin_c4 = []
for e in range(0,len(hist3)):                           #loops over the array to make list
    lhist3.append(hist3[e])
    lbin_c3.append(bin_c3[e])

for r in range(0,len(hist4)):
    lhist4.append(hist4[r])
    lbin_c4.append(bin_c4[r])
    
    
    
ar = list(zip(lbin_c3,lhist3))          #takes the list and makes them ordered pairs
ar1 = list(zip(lbin_c4,lhist4))

ar = [tt for tt in ar if tt[1] > 0]     #filters all of the zeros out 
ar1 = [pp for pp in ar1 if pp[1] > 0]

arx = np.array(ar)[:,0]              # makes the filtered list an array and slects x vals    
ary = np.array(ar)[:,1]              #slects yvals
arys = savgol_filter(ary, 51, 3)     #smooths yvals

ar1x = np.array(ar1)[:,0]
ar1y = np.array(ar1)[:,1]
ar1ys = savgol_filter(ar1y, 51, 3)


plt.figure(figsize=(8,8))
plt.xlim(0,-5E-7)
plt.plot(arx,ary,color='r',alpha=.2)
plt.plot(arx,arys,color='r',label='Cs137')
plt.plot(ar1x,ar1y,color='g',alpha=.2)
plt.plot(ar1x,ar1ys,color='g',label='background')

plt.legend(bbox_to_anchor=(.681, 1), loc=2, borderaxespad=0.,fontsize=18)
plt.ylim(0,200)
plt.xlim(0,-7.5E-7)
plt.xlabel("Area", fontsize= 18)
plt.ylabel("No. of Events", fontsize= 18)
plt.suptitle("Area Distribution", size=22)
plt.show()
```

```python
#    Pulse hight histogram 
#        Have the data ready to feed in, just change the names
#        Pulseht1 and PulsehtBG.
#
#

hist, bin_edges = np.histogram(Pulseht1,bins=4000)     #makes the histogram
bin_c = (bin_edges[:-1]+bin_edges[1:])/2               #finds the centers of the bins

hist1, bin_edges1 = np.histogram(PulsehtBG,bins=4000)
bin_c1 = (bin_edges1[:-1]+bin_edges1[1:])/2

hist0, bin_edges0 = np.histogram(Pulse,bins=4000)     #makes the histogram
bin_c0 = (bin_edges0[:-1]+bin_edges0[1:])/2   

lhist = []                                             #defins the list for the loop
lbin_c = []

lhist1 = []
lbin_c1 = []

lhist0 = []
lbin_c0 = []
for q in range(0,len(hist)):                           #loops over the array to make list
    lhist.append(hist[q])
    lbin_c.append(bin_c[q])

for w in range(0,len(hist1)):
    lhist1.append(hist1[w])
    lbin_c1.append(bin_c1[w])
    
for y in range(0,len(hist0)):
    lhist0.append(hist0[y])
    lbin_c0.append(bin_c0[y])
    
    
    
ph = list(zip(lbin_c,lhist))          #takes the list and makes them ordered pairs
ph1 = list(zip(lbin_c1,lhist1))
ph0 = list(zip(lbin_c0,lhist0))

ph = [t for t in ph if t[1] > 0]     #filters all of the zeros out 
ph1 = [p for p in ph1 if p[1] > 0]
ph0 = [k for k in ph0 if k[1] > 0]


phx = np.array(ph)[:,0]              # makes the filtered list an array and slects x vals    
phy = np.array(ph)[:,1]              #slects yvals
phys = savgol_filter(phy, 21, 3)     #smooths yvals

ph1x = np.array(ph1)[:,0]
ph1y = np.array(ph1)[:,1]
ph1ys = savgol_filter(ph1y, 21, 3)

ph0x = np.array(ph0)[:,0]
ph0y = np.array(ph0)[:,1]
ph0ys = savgol_filter(ph0y, 21, 3)


plt.figure(figsize=(10,10))
plt.xlim(0,-.7)
plt.plot(phx,phy,color='r',alpha=.4)
plt.plot(phx,phys,color='r',label='Cs137')
plt.plot(ph1x,ph1y,color='g',alpha=.4)
plt.plot(ph1x,ph1ys,color='g',label='background')
plt.plot(ph0x,ph0y,color='k',alpha=.4)
plt.plot(ph0x,ph0ys,color='k',label='Cs137 2')


plt.legend(bbox_to_anchor=(.681, 1), loc=2, borderaxespad=0.,fontsize=18)

plt.axvline(-.4, color='b', linestyle=':')
plt.text(-.22,300,'Compton edge 470Kev',rotation=0,fontsize=15,color='b')
plt.axvline(-.54, color='k', linestyle=':')
plt.text(-.56,120,'Cs137 662Kev',rotation=0,fontsize=15,color='k')

plt.xlabel("Pulse Height", fontsize= 18)
plt.ylabel("No. of Events", fontsize= 18)
plt.suptitle("Pulse Height Distribution", size=22)

#plt.savefig('cs137prelim.pdf')


plt.show()
```
