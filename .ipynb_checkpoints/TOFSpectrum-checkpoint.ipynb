```python
import pylab
import numpy
%pylab inline
```

```python
Mobilities=     [1.83*1e-4,1.88*1e-4,2.08*1e-4,2.18*1e-4]
Yields=[1,1e-3,1e-5,1e-3]
DiffusionConsts=[]
Species=        [r'$\mathrm{Ar}_2^{+}$',r'$\mathrm{Ba}^{+}$',r'$\mathrm{Ba}^{++}$','$\mathrm{Cu}^{+}$']
```

```python

kTe=1.38*pow(10,-23)/1.6e-19*300
for i in Mobilities:
    DiffusionConsts.append(i*kTe)
```

```python
AllVels=[]
AllPres=[]
AllE=[]
```

```python
pylab.figure(figsize=(8,3))

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))/(sig*sqrt(2*3.142))


vars=np.arange(0,10,0.0001)
Length=0.2
EField=200*100
Pressure=10.
Totals=zeros_like(vars)

for i in range(0,len(Mobilities)):
    Velocity=Mobilities[i]*EField/Pressure
    TOF=Length/Velocity
    Spread=numpy.sqrt(DiffusionConsts[i]*Length/Pressure)/pow(Velocity,3./2)
    
    pylab.plot(vars,Yields[i]*gaussian(vars,TOF,Spread),label=Species[i])
    Totals+=Yields[i]*gaussian(vars,TOF,Spread)

AllVels.append(np.array(Mobilities)*(EField/Pressure))
AllPres.append(Pressure)
AllE.append(EField)

pylab.semilogy()
pylab.ylim(1e-4,1000)
pylab.xlim(0.35,0.65)
pylab.xlabel("Time of flight / s")
pylab.ylabel("Ion yield (arb)")
pylab.legend(loc='upper left')

pylab.title('E = 200 V/cm, Argon carrier gas, P = 10 bar')
pylab.savefig("Ar200_10bar.pdf")

pylab.show()

```
