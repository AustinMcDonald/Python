```python
Ar10mm   = pd.read_csv('10mmArGap.csv',header=None).values
Ar05mm   = pd.read_csv('5mmArGap.csv',header=None).values
Ar05HDPE = pd.read_csv('5mmArHDPE.csv',header=None).values
Xe05mm   = pd.read_csv('5mmXeGap.csv',header=None).values
Xe05HDPE = pd.read_csv('5mmXeHDPE.csv',header=None).values
Xe85HDPE = pd.read_csv('8.5HDPEXe.csv',header=None).values
Ar85HDPE = pd.read_csv('8.5mmHDPEAr.csv',header=None).values
Ar98HDPE = pd.read_csv('98mmHDPEAr.csv',header=None).values
errs     = pd.read_csv('84HDPE3err.csv',header=None).values
```

```python
def ME(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.mean(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
def SD(array):
    Leng = len(array)
    aa = []
    for x in range(0,Leng):
        qw = np.std(array[x][1:])
        aa.append(qw)
    aa = np.array(aa)
    return aa
```

```python
ER = SD(errs)
```

```python
ME(Ar05mm)
```

```python
plt.figure(figsize=(12,8))
siz = 60
plt.errorbar(Ar10mm[:,0],ME(Ar10mm),yerr=SD(Ar10mm),xerr=0.125,label='Argon 10mm gap',color='r',fmt='o',alpha=0.6,markersize=7)
plt.errorbar(Ar05mm[:,0],ME(Ar05mm),yerr=SD(Ar05mm),xerr=0.125,label='Argon 5mm gap',color='b' ,alpha=0.6,fmt='^',markersize=7)

plt.errorbar(Xe05mm[:,0],ME(Xe05mm),yerr=SD(Xe05mm),xerr=0.125,label='Xenon 5mm gap',color='darkorchid' 
             ,alpha=0.7,fmt='D',markersize=7)



plt.ylabel("Breakdown Voltage / kV",fontsize=26)
plt.xlabel("Pressure / bar",fontsize=26)
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=5, width=1, which='minor')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='upper left',fontsize=24,shadow=True, fancybox=True)
#plt.ylim(0,1000)
plt.xlim(0,11)
plt.grid(True)
plt.tight_layout()

#plt.savefig('/Users/austinmcdonald/Desktop/GaussTest.png')
plt.show()
```
