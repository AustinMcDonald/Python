```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os
import glob
import pandas as pd
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
from scipy import integrate

%matplotlib inline
```

```python
alldata = None
alldata = []
for files in glob.glob("*.csv"):
    alldata.append(files)
#alldata
```

```python
data = dict()
dataf= dict()
area1hold = dict()
area2hold = dict()
areach1 = []
areach2 = []
datacut = dict()
intensity = dict()
leng = len(alldata)
for x in range(0,leng):
    data[x] = pd.read_csv(alldata[x],delimiter=",",names = ['Time','CH1','CH2'],skiprows=22,skipfooter=0,engine='python')
    data[x] = data[x].astype(float)
    data[x]['sig'] = data[x]['CH2']- data[x]['CH1']
    
    data[x]['dif'] = data[x]['CH1'].diff().abs()
    data[x]['dif'] = data[x]['CH2'].diff().abs()
    dataf[x] = data[x][data[x]['dif']<.0002]
    area1hold[x] = integrate.trapz(dataf[x]['CH1'], dataf[x]['Time'])
    area2hold[x] = integrate.trapz(dataf[x]['CH2'], dataf[x]['Time'])

    areach1.append(area1hold[x])
    areach2.append(area2hold[x])

    #data[x]['dif'] = data[x]['sig'].diff().abs()
    #data[x]['BS'] = (data[x]['sig']-backg['sig'])
    #dataf[x] = data[x][data[x]['dif']<.02]
    #datacut[x] = dataf[x][(dataf[x]['wave']>490) & (dataf[x]['wave']<700)]
```

```python
area1 = np.array(areach1)
area1 = area1[area1<1000]
area2 = np.array(areach2)
area2 = area2[area2<1000]

plt.hist(area2,15, color='r')
plt.hist(area1,15, color='g')

plt.show()
```
