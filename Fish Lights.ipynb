```python
led430 = pd.read_csv('430.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
led625 = pd.read_csv('625.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
led660 = pd.read_csv('660.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
ledblu = pd.read_csv('blue.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
ledwht = pd.read_csv('white.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
john   = pd.read_csv('john-filt.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
aust   = pd.read_csv('spect-umfiltered.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
```

```python
fig, ax = plt.subplots(figsize=(15,10))

a,b = john['wave'],john['sig']
plt.plot(a,b/max(b),label= 'Spectrum',linewidth=2)
a,b = led430['wave'],led430['sig']
plt.plot(a,b/max(b)*.7,label= '430 led',linewidth=2)
a,b = led625['wave'],led625['sig']
plt.plot(a,b/max(b)*.52,label= '625 led',linewidth=2)
a,b = led660['wave'],led660['sig']
plt.plot(a,b/max(b)*.36,label= '660 led',linewidth=2)
a,b = ledblu['wave'],ledblu['sig']
plt.plot(a,b/max(b)*.73,label= 'blue led',linewidth=2)
a,b = ledwht['wave'],ledwht['sig']
plt.plot(a,b/max(b)*.68,label= 'white led',linewidth=2)


plt.ylabel("Intensity / arb",fontsize=26)
plt.xlabel("Wavelength / nm",fontsize=26)
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=5, width=1, which='minor')
xmajor_ticks = np.arange(300, 700, 100)
xminor_ticks = np.arange(300, 700, 50)
ymajor_ticks = np.arange(0, 1.1, 0.1)
yminor_ticks = np.arange(0, 1.1, 0.05)
ax.tick_params(axis = 'x', which = 'major', direction='out',width=2,length=6,top="off",right='off')
ax.tick_params(axis = 'x', which = 'minor', direction='out',width=1,length=3,top="off",right='off')
ax.tick_params(axis = 'y', which = 'major', direction='in',width=2,length=6,top="off",right='off')
ax.tick_params(axis = 'y', which = 'minor', direction='in',width=2,length=4,top="off",right='off')
ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor = True)
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor = True)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='upper right',fontsize=24,shadow=True, fancybox=True)
plt.ylim(0,1.01)
plt.xlim(350,700)
plt.grid(True)
plt.tight_layout()

#plt.savefig('/Users/austinmcdonald/Desktop/johnspec.png')

plt.show()
```
