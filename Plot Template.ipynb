```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
Gauss = np.random.normal(100, 5, 10000)
GaussTwo = np.random.normal(70, 15, 10000)
```

```python
plt.figure(figsize=(8,8))
plt.hist(Gauss,bins=np.linspace(0,200,200),label='Gauss',color='b')

plt.hist(GaussTwo,bins=np.linspace(0,200,200),label='Gauss2',color='r',alpha=0.6)


plt.ylabel("Y label",fontsize=26)
plt.xlabel("X label",fontsize=26)
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=5, width=1, which='minor')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='upper right',fontsize=24,shadow=True, fancybox=True)
plt.ylim(0,1000)
plt.xlim(0,200)
plt.grid(True)
plt.tight_layout()

#plt.savefig('/Users/austinmcdonald/Desktop/GaussTest.png')
plt.show()
```

```python
nbins = np.linspace(0,200,200)

hist_Gauss, bin_edges_Gauss = np.histogram(Gauss , nbins)
bin_centres_Gauss = (bin_edges_Gauss[:-1] + bin_edges_Gauss[1:])/2

hist_GaussTwo, bin_edges_GaussTwo = np.histogram(GaussTwo , nbins)
bin_centres_GaussTwo = (bin_edges_GaussTwo[:-1] + bin_edges_GaussTwo[1:])/2
```

```python
plt.figure(figsize=(8,8))

plt.plot(bin_centres_Gauss,hist_Gauss,label='Gauss',color='b',marker='*',linewidth=0,markersize=15,alpha=0.6)

plt.plot(bin_centres_GaussTwo,hist_GaussTwo,label='GaussTwo',color='r',linewidth=2,alpha=0.4)

plt.ylabel("Y label",fontsize=26)
plt.xlabel("X label",fontsize=26)
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=5, width=1, which='minor')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='center right',fontsize=24,shadow=True, fancybox=True)
plt.ylim(0,1000)
plt.xlim(0,220)
plt.grid(True)

plt.text(50, 910, 'Austin was here!',fontsize=24)
plt.axhline(900,linewidth=2,color='m')
plt.axvline(150,linewidth=3,color='orange',alpha=0.9,linestyle='--')

plt.tight_layout()

#plt.savefig('/Users/austinmcdonald/Desktop/GaussTest.png')
plt.show()
```
