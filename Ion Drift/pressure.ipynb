```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tt = [7.3e-6,3.4e-5,5.9e-5,8.3e-5,1.1e-4,1.34e-4,1.6e-4,1.86e-4,2.12e-4,2.39e-4,2.66e-4,2.92e-4,3.22e-4,3.5e-4,3.79e-4,
4.08e-4,4.27e-4,4.67e-4,4.97e-4,5.26e-4,5.56e-4,5.88e-4,6.2e-4,6.49e-4,6.81e-4,7.13e-4,7.45e-4,7.78e-4,8.1e-4,8.43e-4,8.77e-4]
P = np.array(tt)
x1 = np.arange(0,930,30)


ttt = [1.01e-3,1.15e-3,1.27e-3,1.41e-3,1.54e-3,1.68e-3,1.81e-3,1.94e-3,2.11e-3,2.25e-3,2.39e-3,2.54e-3,2.68e-3,2.82e-3
      ,2.95e-3,3.09e-3,3.22e-3]
PP = np.array(ttt)
x2 = np.arange(1020,3000,120)
```

```python
print(len(x2))
print(len(ttt))
```

```python
plt.figure(figsize=(9,9))
plt.semilogy(x1,P,linewidth=0)
plt.scatter(x1/60,P,s=40,color='k')

plt.semilogy(x2,PP,linewidth=0)
plt.scatter(x2/60,PP,s=40,color='k')

plt.xlim(-10/60,3000/60)
plt.ylim(0,5e-3)
plt.grid()
plt.tight_layout()
plt.xlabel(r'Time / min',fontsize=26)
plt.ylabel('pressure / Torr',fontsize=26)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.semilogy(x1,P,linewidth=0)
plt.scatter(x1/60,P*1.33,s=40,color='k')

plt.semilogy(x2,PP*1.33,linewidth=0)
plt.scatter(x2/60,PP*1.33,s=40,color='k')

plt.xlim(-10/60,3000/60)
plt.ylim(0,5e-3)
plt.grid()
plt.tight_layout()
plt.xlabel(r'Time / min',fontsize=26)
plt.ylabel('pressure / mBar',fontsize=26)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
plt.show()
```
