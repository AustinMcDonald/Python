```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Com(E,theta):
    denom = 1 + E/0.511 * (1-np.cos(theta))
    Ep = E/denom
    return Ep

def ComEle(E,theta):
    CON = 2*E/0.511
    THE = 1 - np.cos(theta)
    Ep = E*(CON*THE)/(1+CON*THE)
    return Ep


def KleinNishina(E,theta):
    K0 = E/0.511
    K  = Com(E,theta)/0.511
    KN = (K/K0)**2*(K/K0+K0/K-np.sin(theta)**2)
    return KN
    
```
