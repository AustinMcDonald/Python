```python
noz = np.loadtxt('no-zero-11835.txt')
nof = np.loadtxt('no-forty-12373.txt')
noe = np.loadtxt('no-eighty-12009.txt')
pbe = np.loadtxt('Pb-eighty-12009.txt')
pbf = np.loadtxt('Pb-forty-11342.txt')
pbz = np.loadtxt('Pb-zero-11835.txt')
```

```python
binss = np.linspace(1e-6,.6e-3,100)

hpbz, bin_edges = np.histogram(pbz , bins=binss)
hpbf, bin_edges = np.histogram(pbf , bins=binss)
hpbe, bin_edges = np.histogram(pbe , bins=binss)
hnoz, bin_edges = np.histogram(noz , bins=binss)
hnof, bin_edges = np.histogram(nof , bins=binss)
hnoe, bin_edges = np.histogram(noe , bins=binss)

bin_centres = (bin_edges[:-1] + bin_edges[1:])/2
```

```python
plt.scatter(bin_centres,hpbz/11.835,color='b',label='zero')
plt.scatter(bin_centres,hpbf/11.342,color='r',label='forty')
plt.scatter(bin_centres,hpbe/12.009,color='g',label='eighty')

plt.xlim(1e-6,.6e-3)
plt.legend(loc='upper right')
#plt.ylim(0,150)
plt.show()
```
