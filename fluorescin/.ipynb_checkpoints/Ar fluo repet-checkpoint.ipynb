```python
backg = pd.read_csv('CCD-noise.csv',delimiter=",",names = ['wave','back'],skiprows=33,skipfooter=1,engine='python')
F1 = pd.read_csv('FLUORESCIN-5-1.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
F5 = pd.read_csv('FLUORESCIN-5-2.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
F10 = pd.read_csv('FLUORESCIN-5-3.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
#F0 = pd.read_csv('FLUORESCIN-0.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
```

```python
F1['dif'] = F1['sig'].diff().abs()
F1['BS'] = F1['sig']-backg['back']

F5['dif'] = F5['sig'].diff().abs()
F5['BS'] = F5['sig']-backg['back']

F10['dif'] = F10['sig'].diff().abs()
F10['BS'] = F10['sig']-backg['back']

#F0['dif'] = F0['sig'].diff().abs()
#F0['BS'] = F0['sig']-backg['back']

```

```python
plt.figure(figsize=(10,10))

#F0f = F0[F0['dif']<.1]
F1f = F1[F1['dif']<.1]
F5f = F5[F5['dif']<.1]
F10f = F10[F10['dif']<.1]


#plt.plot(F0f['wave'],F0f['BS'],color=cm.copper(0/4),label=r'$0 \mu M$'+' Fluorescein',linewidth=2)
plt.plot(F1f['wave'],F1f['BS'],color=cm.copper(1.5/4),label=r'$1 \mu M$'+' Fluorescein',linewidth=2)
plt.plot(F5f['wave'],F5f['BS'],color=cm.copper(2.5/4),label=r'$5 \mu M$'+' Fluorescein',linewidth=2)
plt.plot(F10f['wave'],F10f['BS'],color=cm.copper(3/4),label=r'$10 \mu M$'+' Fluorescein',linewidth=2)

#plt.text(590, .65, r'$\frac{F_{max}}{F_{min}} = 11.69$', fontsize=25)

plt.xlabel(r'Wavelength ($nm$)', fontsize=32)
plt.ylabel('Relative Intensity', fontsize=32)
#plt.title('Fluorescein Emission Spectrum ', fontsize=32)
plt.legend(bbox_to_anchor=(1,1), loc=1, borderaxespad=0.,fontsize=22,shadow=True, fancybox=True)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xticks([450,475,500,525,550,575,600,625,650])
plt.tight_layout()

plt.xlim(450,650)
plt.ylim(0,.98)
#plt.savefig('/Users/austinmcdonald/Desktop/Plots/paper8-16/Ar_Fluo_IF.pdf')

plt.show()
```
