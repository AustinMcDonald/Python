```python
dry = pd.read_csv('dry1.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
clean = pd.read_csv('clean1.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
```

```python
dry['dif'] = dry['sig'].diff().abs()
dry = dry[dry['dif']<.005]
dry['BS'] = dry['sig']-clean['sig']
dryf = dry[dry['dif']<.005]

clean['dif'] = clean['sig'].diff().abs()
clean = clean[clean['dif']<.005]
```

```python
plt.figure(figsize=(8,8))
plt.plot(dryf['wave'],dryf['BS'])
plt.title('dryed Fluorescein - fiber background ', fontsize=20)

plt.xlim(450,650)
plt.ylim(0,0.4)
#plt.savefig('/Users/austinmcdonald/Desktop/Plots/paper8-16/dryedtest_signal.pdf')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(dry['wave'],dry['sig'],label='Dry Fluorescein')
plt.plot(clean['wave'],clean['sig'],label='Cleaned fiber')
plt.legend(bbox_to_anchor=(1,1), loc=1, borderaxespad=0.,fontsize=22,shadow=True, fancybox=True)
plt.title('dryed Fluorescein and fiber background ', fontsize=20)


plt.xlim(450,650)
plt.ylim(0,0.5)
#plt.savefig('/Users/austinmcdonald/Desktop/Plots/paper8-16/dryedtest_both.pdf')
plt.show()
```
