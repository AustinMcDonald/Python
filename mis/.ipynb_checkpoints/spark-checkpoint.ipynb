```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os
import glob
import pandas as pd
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
from scipy import integrate
%matplotlib inline
```

```python
#  This seems to be inconsistant leaving it here for refrence
#
def f(t):
    return (.036*t)/(np.log(.00038*t))

def g(t):
    return (.0195*t)/(np.log(.000376*t))

def h(t):
    return (.03059*t)/(np.log(.0002166*t))

def i(t):
    return (.004*t)/(np.log(.0002666*t))

t1 = np.arange(2700, 250000, 100)
t2 = np.arange(2700, 250000, 100)
t3 = np.arange(4650, 250000, 100)
t4 = np.arange(3760, 250000, 100)

plt.figure(figsize=(8,8))
plt.loglog(t1,f(t1), label = 'Air')
plt.loglog(t2,g(t2), label = 'Xenon')
plt.loglog(t3,h(t3), label = 'Nitrogen')
plt.loglog(t4,i(t4), label = 'Argon')
plt.grid(True, which="both")
plt.xlabel('Pressure times Distance in mm mTorr')
plt.ylabel('Breakdown Voltage')
plt.title('Parallel Plate breakdown voltage')
plt.xlim(0,500000)
plt.ylim(0,2000)
plt.legend( )
plt.show()
```

```python
# This is Jordans measurements in the calbration stand finished on 12-22-16
spark = pd.read_csv('sparkgap.csv',delimiter=",",names = ['PD','Volt','Err'],skiprows=0,skipfooter=0,engine='python')
```

```python
#plt.figure(figsize=(8,8))
yerr = spark['Err']

plt.loglog(spark['PD'],spark['Volt'], '.',color = 'b')
plt.errorbar(spark['PD'],spark['Volt'],yerr=yerr, linestyle="None", color='r',label=None)
plt.grid(True, which="both")
plt.xlabel('pd in cm Torr')
plt.ylabel('Breakdown Voltage')
plt.title('Jordans needle breakdown voltage')
plt.xlim(0,200)
plt.ylim(100,2200)
plt.legend(loc='center right' )
plt.show()
```

```python
# V = B*pd/( ln(A*pd)+ln(ln(1+1/gamma)) )
#
ArA = 12    #per Torr cm
ArB = 180     #per Torr cm
gammaAr = 0.01
def VAr(pd):
    return (ArB*pd)/(np.log(ArA*pd)+np.log(np.log(1+1/gammaAr)))
t1 = np.arange(1e-4, 1000, 0.005)
plt.figure(figsize=(14,8))

yerr = spark['Err']

plt.loglog(spark['PD'],spark['Volt'], '.',color = 'r')
plt.errorbar(spark['PD'],spark['Volt'],yerr=yerr, linestyle="None", color='r',label = None)

plt.loglog(t1,VAr(t1), label = 'Ar')

plt.grid(True, which="both")
plt.xlabel('pd in cm Torr')
plt.ylabel('Voltage')
plt.title('Breakdown Voltage')
plt.xlim(1e-2,1e3)
plt.ylim(0,2000)
plt.legend(loc='center right' )
plt.show()
```
