```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

def FIND_MAX(X,Y):
    TriggerTime = find_nearest(X,0)
    Trigger     = np.where(X==TriggerTime)[0][0]+5
    xx = moving_average(X,20)
    yy = moving_average(Y,20)
    
    xxx = xx[Trigger:]
    yyy = yy[Trigger:]
    
    maxy = max(yyy)
    
    hold      = find_nearest(yyy,maxy)
    peak     = np.where(yyy==hold)[0][0]

    xxx = xx[Trigger:Trigger+peak]
    yyy = yy[Trigger:Trigger+peak]
    
    hold      = find_nearest(yyy,maxy*0.1)
    edge     = np.where(yyy==hold)[0][0]
    return Trigger+peak, Trigger+edge

def FIND_MIN(X,Y):
    TriggerTime = find_nearest(X,0)
    Trigger     = np.where(X==TriggerTime)[0][0]+20
    xx = moving_average(X,20)
    yy = moving_average(Y,20)
    
    xxx = xx[Trigger:]
    yyy = yy[Trigger:]
    
    miny = min(yyy)
    
    hold      = find_nearest(yyy,miny)
    MINLOC     = np.where(yyy==hold)[0][0]
    
    TT = Trigger+MINLOC
    xxxx = xx[Trigger:TT]
    yyyy = yy[Trigger:TT]
    
    hold      = find_nearest(yyyy,miny*0.1)
    REDGE     = np.where(yyyy==hold)[0][0]
    return MINLOC+Trigger,miny,REDGE+Trigger

def func2(t,t0,sig,a):
    tau = 120
    A = (sig**2-2*(t-t0)*tau)/(2*tau**2)
    B = (-sig**2 +(t-t0)*tau)/(np.sqrt(2.)*tau*sig)
    C = np.exp(A)
    D = (1 + erf(B))
    E = a*C*D
    return E
```

```python


def gaus(x,x0,sigma,a):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

def squa(x,x0,width):
    #width = width/2
    y = np.zeros(len(x))
    L1 = find_nearest(x, x0-width)
    L2 = np.where(x==L1)[0][0]
    L3 = find_nearest(x, x0+width)
    L4 = np.where(x==L3)[0][0]
    y[L2:L4] = 1
    return y
```

```python
from scipy import optimize
```

```python

tau= 120
x = np.linspace(0,1000,50e3)

sig = 1
w = sig/0.68 
ys = squa(x,10,w)
ys = ys/np.sum(ys)
yg = gaus(x,10,sig,1) 
yg = yg/np.sum(yg)

c = (erf(x)+1e-6)*np.exp(-x/tau)

plt.plot(x,ys,lw=2,label="Square")
plt.plot(x,yg,lw=2,label="Gaussian")
plt.axvline(10-sig,color='r')
plt.axvline(10+sig,color='r')

plt.xlim(0,20)
plt.legend(loc="upper right")
plt.show()

plt.plot(x,c)
plt.xlim(0,20)
plt.show()

plt.figure(figsize=(7,7))
yc = signal.convolve(ys, c, mode='full') / c.sum()
ydc, remainder = signal.deconvolve(yc, c)
ydc *= c.sum()
plt.plot(x[0:5000], yc[0:5000],lw=2,label="Square",alpha=0.6)
optim,eh = optimize.curve_fit(func2,x[0:5000], yc[0:5000])
print(optim)
plt.plot(x[0:5000],func2(x[0:5000],*optim),color='r',linestyle=':')


yc = signal.convolve(yg, c, mode='full') / c.sum()
ydc, remainder = signal.deconvolve(yc, c)
ydc *= c.sum()
plt.plot(x[0:5000], yc[0:5000],lw=2,label="Gaussian",alpha=0.6)
optim,eh = optimize.curve_fit(func2,x[0:5000], yc[0:5000])
print(optim)
plt.plot(x[0:5000],func2(x[0:5000],*optim),color='r',linestyle=':')

plt.xlim(0,20)

plt.legend(loc="lower right")
plt.show()
```
