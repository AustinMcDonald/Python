```python
def DeltaM(ml):
    return (ml**2-7.5e-5)**(1/2)
def DeltaMup(ml):
    return (ml**2-8.09e-5)**(1/2)
def DeltaMdn(ml):
    return (ml**2-7.03e-5)**(1/2)
```
